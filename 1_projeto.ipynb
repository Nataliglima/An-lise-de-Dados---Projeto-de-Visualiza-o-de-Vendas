```python
import matplotlib.pyplot as plt

# Dados dos cursos e quantidades vendidas
dados = [
    ("Arquitetura de Microserviços", 30),
    ("Banco de Dados SQL e NoSQL", 30),
    ("Cloud Computing com AWS", 37),
    ("Desenvolvimento Mobile com React Native", 63),
    ("Desenvolvimento Web com HTML e CSS", 75),
    ("DevOps: Integração e Entrega Contínua", 14),
    ("Introdução ao Machine Learning", 37),
    ("Introdução à Programação em Python", 95),
    ("JavaScript Avançado: Frameworks e Bibliotecas", 53),
    ("Segurança da Informação: Fundamentos", 15),
]

# Separando os dados em listas de nomes dos cursos e total de vendas
nomes_cursos = [curso[0] for curso in dados]
total_vendas = [curso[1] for curso in dados]

# Encontrando o curso com o maior número de vendas
max_vendas = max(total_vendas)
curso_mais_vendido = nomes_cursos[total_vendas.index(max_vendas)]

# Criando o gráfico de barras
plt.figure(figsize=(12, 8))
bars = plt.bar(nomes_cursos, total_vendas, color='skyblue')

# Destacando o curso mais vendido
bars[total_vendas.index(max_vendas)].set_color('orange')

# Adicionando anotações para o curso mais vendido
plt.text(total_vendas.index(max_vendas), max_vendas + 2,
         f'{curso_mais_vendido}\nVendas: {max_vendas}', ha='center', color='black', weight='bold')

plt.xlabel('Cursos')
plt.ylabel('Quantidade Total de Vendas')
plt.title('Quantidade Total de Vendas por Curso')
plt.xticks(rotation=90)
plt.tight_layout()

plt.show()
```

```python
df = pd.read_csv('/content/dados_cursos.csv')

# Agrupando por data para somar as vendas diárias
vendas_diarias = df.groupby('Data')['Quantidade de Vendas'].sum().reset_index()

# Plotando o gráfico de linha
plt.figure(figsize=(12, 6))
plt.plot(vendas_diarias['Data'], vendas_diarias['Quantidade de Vendas'], marker='o')
plt.title('Distribuição das Vendas ao Longo do Tempo')
plt.xlabel('Data')
plt.ylabel('Quantidade de Vendas')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
